```python
import os

if __name__ == "__main__":
    cwd = os.getcwd()
    # root_dir = "./c\my001\.vs\my001\FileContentIndex"
    root_dir = cwd
    for (root, dirs, files) in os.walk(root_dir):
        print("# root : " + root)
        if len(dirs) > 0:
            for dir_name in dirs:
                print("dir: " + dir_name)

        if len(files) > 0:
            for file_name in files:
                print("file: " + file_name)
```
