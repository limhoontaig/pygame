```python
import os
import re
import pathlib
import stat
from datetime import datetime
from PIL import Image
from PIL.ExifTags import TAGS
from time import time
import time

```

```python
path = 'C:\\source\\pygame\\Nado Game\\pyqt5\\사진관리\\data'
#path = 'D:\\사진\\2006년\\060116 채혁입대'

def takePictureTime(path, f):
    filename, fileExtension = os.path.splitext(f)
    if fileExtension == 'db':
        return
    else:
        filename = pathlib.Path(path, f)
        #try :
        with Image.open(filename) as im:
            info = im._getexif()
        # 새로운 딕셔너리 생성
        taglabel = {}
        for tag, value in info.items():
            decoded = TAGS.get(tag, tag)
            taglabel[decoded] = value
        print(f, taglabel)
        try:
            s = taglabel['DateTimeOriginal']
            # print(s)
            timestamp = time.mktime(datetime.strptime(s, '%Y:%m:%d %H:%M:%S').timetuple())
            return timestamp
        except:
            return
    
    #except:
    #    now = datetime.now()
    #    timestamp = datetime.timestamp(now)
    #return timestamp


files = os.listdir(path)
for file in files:
    print(takePictureTime(path, file))
    if takePictureTime(path, file):
        min_time = takePictureTime(path, file)
    else:        
        f = pathlib.Path(path, file)
        T = os.stat(f)
        c_time = T.st_ctime
        m_time = T.st_mtime
        a_time = T.st_atime
        min_time = min(c_time, m_time, a_time)
    
    print(min_time)
    dt = datetime.fromtimestamp(min_time)
    y = dt.strftime("%Y%m%d_%H%M%S")
    new_f = y+'_'+file
    print(y, file, new_f)
    

text = ['891031', '1923-12-11 전주']
checker = re.compile(r'^(19\d\d|20\d\d|\d\d)[년\-_. ]?(0[1-9]|1[012])[월\-_. ]?(0[1-9]|[12][0-9]|3[01])[일]?')
for t in text:
    D = checker.search(t)
    if D:
        print(D.end())
    else:
        print('None')
        
    


               
```
